```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

features = ["enginesize", "horsepower", "curbweight", "citympg"]

for i, feat in enumerate(features):
    row, col = i // 2, i % 2
    axes[row][col].scatter(
        df[feat], df["price"],
        alpha=0.5,
        color="#378ADD",
        edgecolors="none"
    )
    axes[row][col].set_xlabel(feat)
    axes[row][col].set_ylabel("Price")
    axes[row][col].set_title(f"{feat} vs Price")

plt.tight_layout()
plt.show()
```

--------------------------------------------------Data Cleaning--------------------------------------------------------

```python
df['brand'] = df['CarName'].str.split().str[0].str.lower()
typo_map = {
    "vw": "volkswagen",
    "vokswagen": "volkswagen",
    "toyouta": "toyota",
    "maxda": "mazda",
    "porcshce": "porsche"
}

df["brand"] = df["brand"].replace(typo_map)
df = df.drop('CarName', axis = 1)
print(df["brand"].value_counts())
```

```python
# Cell 8 — Drop useless columns
df = df.drop("car_ID", axis=1)
print("Shape after dropping car_ID:", df.shape)
# Should show (205, 25)
```

```python
cylinder_map = {
    "two":2,
    "three":3,
    "four":4,
    "five":5,
    "six":6,
    "eight":8,
    "twelve":12    
}

df["cylindernumber"] = df["cylindernumber"].map(cylinder_map)

print(df["cylindernumber"].value_counts())
print("dtype:",df["cylindernumber"].dtype)
```

```python
# Cell 10 — Encode text columns
print("Object columns before encoding:")
print(df.select_dtypes(include="object").columns.tolist())

df_encoded = pd.get_dummies(df, drop_first=True, dtype=int)

print("\nShape before encoding:", df.shape)
print("Shape after encoding: ", df_encoded.shape)
print("\nNew columns added:", df_encoded.shape[1] - df.shape[1])
```

```python
df_feat = pd.read_csv("../data/Car_data.csv")

df_feat["hp_per_litre"] = df_feat["horsepower"] / (df_feat["enginesize"] + 1)
df_feat["weight_per_hp"] = df_feat["curbweight"] / (df_feat["horsepower"] + 1)
df_feat["mpg_combined"] = (df_feat["citympg"] + df_feat["highwaympg"]) / 2
df_feat["engine_per_weight"] = df_feat["enginesize"] / (df_feat["curbweight"] + 1)

df_feat["car_volume"] = df_feat["carlength"] * df_feat["carwidth"] * df_feat["carheight"]
df_feat["wheelbase_ratio"] = df_feat["wheelbase"] / df_feat["carlength"]

df_feat["log_price"] = np.log1p(df_feat["price"])

df_feat["is_luxury"] = (df_feat["price"] > 20000).astype(int)
df_feat["is_powerful"] = (df_feat["horsepower"] > 140).astype(int)
df_feat["is_fuel_efficient"] = (df_feat["citympg"] > 30).astype(int)

# ---- ADD THESE 3 LINES ----
df_feat = df_feat.drop(columns=["CarName"])        # drop, too many unique values
df_feat = pd.get_dummies(df_feat, drop_first=True) # encode remaining string columns
# ---------------------------

df_feat.to_csv("../data/car_data_features.csv", index=False)
print("Saved with features!", df_feat.shape[1], "total columns")
```
